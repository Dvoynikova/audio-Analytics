```
DATA_PATH = '/content/gdrive/My Drive/Data_Kaggle/'
model1_label = 'baseline_40.csv'
model2_label = 'baseline_38.csv'
model1_path = 'baseline_40_probability.csv'
model2_path = 'baseline_38_probability.csv'

train_meta_fname = 'train.csv'
```

```

model1 = pd.read_csv(os.path.join(DATA_PATH, model1_path))
model2 = pd.read_csv(os.path.join(DATA_PATH, model2_path))
df_train = pd.read_csv(os.path.join(DATA_PATH, train_meta_fname))
```

```
#кодируем лейблы
n_classes = df_train.label.nunique()
print(n_classes)
classes_dict = {cl:i for i,cl in enumerate(df_train.label.unique())}
decoder = {classes_dict[cl]:cl for cl in classes_dict}
```

```
#создаем табличку для объединения
model_merge = pd.DataFrame(columns = ['fname', 'label'])
model_merge['fname'] = model1['fname']
```

```
#высчитываем среднее арифметическое вероятностей
for i in range(len(model_merge)):
  feature1 = model1['label'][i].replace('\n','').replace('[','').replace(']','').replace('         ',' ').replace('     ',' ').replace('   ',' ').replace('  ',' ').lstrip(' ').rstrip(' ').split(' ')
  feature2 = model2['label'][i].replace('\n','').replace('[','').replace(']','').replace('         ',' ').replace('     ',' ').replace('   ',' ').replace('  ',' ').lstrip(' ').rstrip(' ').split(' ')
  
  feature1 = list(map(float, feature1))
  feature2 = list(map(float, feature2))

  feature_merge = list(map(lambda x,y: (x + y )/2, feature1, feature2 ))
  
  model_merge['label'].iloc[i] = decoder.get(feature_merge.index(np.max(feature_merge))) # ищет в decoder по индексу максимального элемента в листе

```
